```python
from data.loader import load_complete_dataset
from data.preprocessing import add_module_columns

dataset = load_complete_dataset()
add_module_columns(dataset)
```

```python
import pandas as pd
from scipy import stats


groups = dataset[["actitivy_label","accelerometer_module","gyroscope_module","magnetometer_module"]].groupby("actitivy_label")
groups = dataset.groupby("actitivy_label")
cols=["accelerometer_module","gyroscope_module","magnetometer_module"]
df_results = pd.DataFrame(columns=["Activity", "Atribute", "D", "p"])
for col in cols:
    for name,group in groups:
        data = group[col]
        D, p = stats.kstest(data, 'norm')
        df_results.loc[len(df_results)] = {
            "Activity": name,
            "Atribute": col,
            "D": D,
            "p": p
        }
df_results
```

```python
from scipy import stats

cols=["accelerometer_module","gyroscope_module","magnetometer_module"]
groups = dataset.groupby("actitivy_label")

df_results = pd.DataFrame(columns=["Module", "H", "P"])
for col in cols:
    samples = [group[col].values for _, group in groups]
    H, p = stats.kruskal(*samples)
    df_results.loc[len(df_results)] = {
        "Module": col,
        "H": H,
        "P": p
    }


df_results
```
